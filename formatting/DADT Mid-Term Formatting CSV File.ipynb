```python
import pandas as pd

listings = pd.read_csv('listings.csv')
df = pd.DataFrame(listings)
df
```

```python
# Remove 2 columns for relevancy
columns_to_drop = ['neighbourhood_group', 'license']
existing_columns = df.columns.tolist()
columns_to_drop = [col for col in columns_to_drop if col in existing_columns]
df = df.drop(columns=columns_to_drop, axis=1)
df
```

```python
# Finding out data types for each column
df.info()
```

```python
# A focus on the 'name' column is required as it contains multiple entries of data which can be seperated into different fields
# That column has information of the location name, number of stars, number of bedrooms, number of beds, number of baths-
# which we to be seperate into different columns for purposes of data preprocessing and standardisation.
df[['name']]
```

```python
# Splitting the 'name' column into its individual attributes and making it thier own column on the dataframe - apprach 1
df[['Location', 'Rating', 'Bedrooms', 'Beds', 'Baths']] = df['name'].str.split('·', expand=True)
df = df.drop(columns=['name'])
df
```

```python
# Cleaning up repeates and order of columns in the dataframe
df = df[['id', 'Location', 'Rating', 'Bedrooms', 'Beds', 'Baths', 'room_type', 'price', 'neighbourhood', 'host_name', 'host_id', 'calculated_host_listings_count', 
        'latitude', 'longitude', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'number_of_reviews_ltm', 'last_review', 'availability_365']]
df
```

```python
# Removing rows with empty fields
# Reason for this is that we want data completeness, which is crucial for further future developement.
df = df.dropna()
df
```

```python
df.to_csv('fmtd-listings.csv', index=False)
```
